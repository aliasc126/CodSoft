```python
import pandas as pd
# read the csv named tested

test = pd.read_csv('tested.csv')
test.head()
```

```python
testing = test.copy()
```

```python
testing.dtypes
```

```python
testing.drop('Name', axis=1, inplace=True)
testing.dtypes
```

```python
testing.dtypes
```

```python
# encode the column Ticket and Cabin and EMbarked
testing['Sex'] = testing['Sex'].replace({'male': 1, 'female': 0})
testing['Ticket'] = testing['Ticket'].astype('category')
testing['Cabin'] = testing['Cabin'].astype('category')
testing['Embarked'] = testing['Embarked'].astype('category')
```

```python
# remove all the null coluns
testing.dropna(inplace=True)
```

```python
testing.head()
```

```python
testing['Cabin'] = testing['Cabin'].str.split()
testing = testing.explode('Cabin')

# Reset the index to get consecutive row numbers
testing = testing.reset_index(drop=True)
```

```python
testing.head()
```

```python
# drop the PassengerId column
testing.drop('PassengerId', axis=1, inplace=True)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
testing['Cabin'] = label_encoder.fit_transform(testing['Cabin'])
testing['Embarked'] = label_encoder.fit_transform(testing['Embarked'])
testing['Ticket'] = label_encoder.fit_transform(testing['Ticket'])
```

```python

X = testing.drop('Survived', axis=1)
y = testing['Survived']

# split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)

# import random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

# fit the model
rfc.fit(X_train, y_train)

# predict the model
y_pred = rfc.predict(X_test)

# import the accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
